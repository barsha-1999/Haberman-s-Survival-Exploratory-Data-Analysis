# Haberman Cancer Survival dataset
---
Haberman's Survival Dataset : https://www.kaggle.com/gilsousa/habermans-survival-data-set

The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.


Objective : To predict Whether a patient will survive 5 years or more after the operation based on age , year of operation and the number of positive axillary nodes.

Attribute Information:

*  Age of patient at the time of operation (numerical)
*  Patient's year of operation (year - 1900, numerical)


*  Number of positive axillary nodes detected (numerical)
* Survival status (class attribute) : 1 = the patient survived 5 years or longer AND 2 = the patient died within 5 years











```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
haberman=pd.read_csv('haberman.csv',names=['Age of Patient',
                                           'Year of Operation',
                                           'Number of PA nodes detected',
                                           'Survival status'])
```

## Obeservations:


1.   This dataset Contains 2 Labels- 1 and 2. Label 1 corresponds to the patient survived 5 years or longer after the surgery and Label 2 corresponds to the patient died within 5 years after the surgery.
2.   Label 1 contains 225 values and Label 2 contains 81 values. So, Its  unbalanced dataset.



# Univariate analysis

```python
Survived=haberman.loc[haberman['Survival status']==1]
Died=haberman.loc[haberman['Survival status']==2]
```

```python
Survived.describe()
```

```python
Died.describe()
```

## Obeservations:

1.  Between 1958 to 1969, 22 patients were survived, and 81 patients who were died within 5 years of surgery. 
2.The minimum age of the patient who was survived was 30 years, and the maximum age was 77 years. Whereas the minimum age of the patient who was died was 34 years and the maximum age was 83 years.
3.In case of died patients, maximum number of positive axillary nodes detected was 52 and in case of survived patients,there was maximum 46 positive nodes detected.





## Histogram with PDF

```python
sns.FacetGrid(haberman,hue="Survival status",size=7) \
   .map(sns.distplot,"Age of Patient") \
   .add_legend()
plt.show()   
```

```python
sns.FacetGrid(haberman,hue="Survival status",size=7) \
   .map(sns.distplot,"Year of Operation") \
   .add_legend()
plt.show()   
```

```python
sns.FacetGrid(haberman,hue="Survival status",size=7) \
   .map(sns.distplot,"Number of PA nodes detected") \
   .add_legend()
plt.show()   
```

## PDF and CDF Plot


```python
counts, bin_edges = np.histogram(Survived['Age of Patient'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.legend(['survive_pdf', 'survive_cdf','die_pdf', 'die_cdf'])
plt.show()

print("***********************************************************************")

counts, bin_edges = np.histogram(Died['Age of Patient'], bins=20, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:] ,cdf)
plt.legend(['survive_pdf', 'survive_cdf','die_pdf', 'die_cdf'])
plt.show()
```

## Observations:
